## 2. Interpreting Data Description

The results show 8 numbers for each column in your original dataset : 

- **Count**: Shows how many rows have non-missing values in each column.
- **Missing Values**: Can occur for various reasons, such as data not being collected (e.g., the size of the 2nd bedroom in a 1 bedroom house).
- **Mean**: The average value for each column.
- **Standard Deviation (std)**: Measures the numerical spread of the values in each column.
- **Percentiles (min, 25%, 50%, 75%, max)**:
    - Min: The smallest value in the column.
    - 25% (25th percentile): A value bigger than 25% of the values and smaller than the remaining 75%.
    - 50% (median or 50th percentile): A middle value that separates the higher half from the lower half.
    - 75% (75th percentile): A value that is greater than 75% of the data.
    - Max: The largest value in the column.
